**CGPA Prediction**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('Cgpa_prediction.xlsx')
df
```

```python
plt.xlabel("Semester")
plt.ylabel("Cgpa")
plt.scatter(df['semester'], df['cgpa'], color = 'red', marker = '+')
```

```python
# applying model
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['semester']], df['cgpa'])
```

```python
reg.predict([[5]])
```
