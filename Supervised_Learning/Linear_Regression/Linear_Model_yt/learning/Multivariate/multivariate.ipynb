Given these home prices find out price of a home that has,
 - 3000 sqr ft area, 3 bedrooms, 40 year old
 - 2500 sqr ft area, 4 bedrooms, 5 year old

 price = m1 * area + m2 * bedrooms + m3 * age + b

```python
df = pd.read_csv('homeprices.csv')
df.head(10)
```

```python
med = df.bedrooms.median()
```

```python
df.bedrooms.fillna(med, inplace=True)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area', 'bedrooms', 'age']], df['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```

```python
reg.predict([[2500, 4.0, 5]])
```

<h2>Practice Exercise</h2>

In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

- 2 yr experience, 9 test score, 6 interview score

- 12 yr experience, 10 test score, 10 interview score

```python
data = pd.read_csv('hiring.csv')
```

```python
data
```

```python
import math
median_value = math.floor(data['test_score(out of 10)'].median())
median_value
```

```python
# data['test_score(out of 10)'].fillna(median_value, inplace=True)

data['test_score(out of 10)'] = data['test_score(out of 10)'].fillna(median_value)
```

```python
data.head(10)
```

```python
data['experience'] = data['experience'].fillna('zero')
```

```python
data
```

```python
def safe_word_to_num(val):
    try:
        if isinstance(val, str):  # Only apply word_to_num if it's a string
            return w2n.word_to_num(val)
        else:
            return val  # Return the value as it is if it's not a string
    except ValueError:
        return np.nan  # In case of conversion error, return NaN

# Apply the function to the 'experience' column
data['experience'] = data['experience'].apply(safe_word_to_num)

# Display the updated DataFrame
data
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(data[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], data['salary($)'])
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg.predict([[12, 10, 10]])
```
