```python
import nltk
import spacy
```

**Stemming in NLTK**

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "eats" ,"eat", "ate", "adjustable", "rafting" ,"ability", "meeting", "better"]

print(f"{"Word".ljust(11)} | {"  Stem word"}")
for word in words:
    print(f"{word.ljust(10)} -> {stemmer.stem(word)}")
```

**Lematization in Spacy**

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("eating eats eat ate adjustable rafting ability meeting better")

for word in doc:
    print(f"{word.text.ljust(10)} -> {word.lemma_}")
```

```python
doc = nlp("Mando talked 3 hours although talking isn't his thing")

for word in doc:
    print(f"{word.text.ljust(10)} | {word.lemma_.ljust(10)} | {word.lemma}")
```

```python
ar = nlp.get_pipe('attribute_ruler')
ar.add([[{"TEXT": "Bro"}],[{"TEXT": "Brah"}]], {"LEMMA": "Brother"})


doc = nlp("Bro, you wanna go? Brah, don't say no I am exhausted")

for token in doc:
    print(token.text, "|", token.lemma_)
```
